```python
# Arcturus Test
LOGG = 1.66
FE_H  = -0.52
TEFF = 4286 
```

```python
gd, = np.where(((np.abs(10**ISO['LOGTE']-(TEFF)) < 200) &
                (np.abs(ISO['LOGG']-LOGG) < 0.005) &
                (np.abs(ISO['FEH']-FE_H) < 0.005) ) )

            # Take the list of isochrone fittings and find the mena, mediant, and
            # deviation for the values they gave back. If they weren't any good
            # firts then make it nan
if len(gd) > 0:
    iso_meanM = np.mean(ISO['MASS'][gd])
    iso_medianM = np.median(ISO["MASS"][gd])
    iso_stdM = np.std(ISO["MASS"][gd])
                
    foo = 10**ISO['LOGL'][gd]
    iso_meanL = np.mean(foo)
    iso_medianL = np.median(foo)
    iso_stdL = np.std(foo)

else:
    iso_meanM = np.nan
    iso_medianM = np.nan
    iso_stdM = np.nan
    iso_meanL = np.nan
    iso_medianL = np.nan
    iso_stdL = np.nan
# Now add on the mass luminocity and radius values
Arc_M = iso_meanM*u.solMass


Arc_L= iso_meanL * u.solLum

Arc_R = np.sqrt( Arc_L / (4 * np.pi * sigma_sb * (TEFF*u.K)**4 )) .to(u.solRad)
print(Arc_L, Arc_M, Arc_R)
```
