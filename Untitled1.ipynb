```python
# First step is to get the data.
from astropy.io import fits
from astropy.table import Table

import os

import numpy as np
import matplotlib.pyplot as plt

AllAvgSimp = Table(fits.getdata('/home/cody/Binary Search/AllAvgSimp.fits',1))
AllVisSimp = Table(fits.getdata('/home/cody/Binary Search/AllVisSimp.fits',1))
AllAverage = Table(fits.getdata('/home/cody/Binary Search/allMC.avg.nocut.fits',1))
AllVisit   = Table(fits.getdata('/home/cody/Binary Search/allVisit-t9-l31c-58104.mc.fits',1))
```

```python
# Similar to the problem we had with radial velocities bad data is stores as -9999.9999 rather than just a nan 
# or something simple like that, so I want to go through and remove any points that give these bad magnitude values.
J_Filter_All = AllAverage['J']
for j in range(len(J_Filter_All)):
    if abs(J_Filter_All[j]) > 900:
        J_Filter_All[j] = 'nan'
J_Filter_All
K_Filter_All = AllAverage['K']
for k in range(len(K_Filter_All)):
    if abs(K_Filter_All[k]) > 900:
        K_Filter_All[k] = 'nan'
        
H_Filter_All = AllAverage['H']
for h in range(len(H_Filter_All)):
    if abs(H_Filter_All[h]) > 900:
        H_Filter_All[h] = 'nan'
        
```

```python
plt.figure(3,figsize=(8.5,11))
plt.scatter(H_Filter_All - K_Filter_All,J_Filter_All)
plt.ylim((6,18))
plt.gca().invert_yaxis()
#plt.ylim((-1.5,20))
#plt.xlim((-1.5,3))
plt.ylabel('H Filter')
plt.xlabel('K-J Filter')
plt.show()
```
